```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>AI List Assist | Developer Architecture Portal</title>
    <script src="https://cdn.tailwindcss.com"></script>
    <script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
    <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;600;700&display=swap" rel="stylesheet">
    <style>
        body { font-family: 'Inter', sans-serif; background-color: #F8FAFC; color: #1E293B; }
        .glass-card { background: rgba(255, 255, 255, 0.95); backdrop-filter: blur(10px); border: 1px solid #E2E8F0; }
        .flow-arrow::after {
            content: ''; position: absolute; right: -20px; top: 50%; transform: translateY(-50%);
            border-width: 8px 0 8px 12px; border-color: transparent transparent transparent #CBD5E1;
        }
        .active-node { border-color: #3B82F6; background-color: #EFF6FF; }
        .chart-container { position: relative; width: 100%; height: 300px; max-width: 600px; margin: 0 auto; }
        /* Custom Scrollbar */
        ::-webkit-scrollbar { width: 8px; }
        ::-webkit-scrollbar-track { background: #F1F5F9; }
        ::-webkit-scrollbar-thumb { background: #CBD5E1; border-radius: 4px; }
        ::-webkit-scrollbar-thumb:hover { background: #94A3B8; }
        .code-block { font-family: 'Fira Code', monospace; }
    </style>
</head>
<body class="min-h-screen flex flex-col">

    <!-- Header -->
    <header class="bg-white border-b border-slate-200 sticky top-0 z-50">
        <div class="max-w-7xl mx-auto px-4 sm:px-6 lg:px-8 h-16 flex items-center justify-between">
            <div class="flex items-center gap-3">
                <div class="w-8 h-8 bg-blue-600 rounded-lg flex items-center justify-center text-white font-bold">AI</div>
                <div>
                    <h1 class="text-lg font-bold tracking-tight text-slate-900">List Assist <span class="text-slate-400 font-normal">| DevPortal</span></h1>
                    <p class="text-xs text-emerald-600 font-medium flex items-center gap-1">
                        <span class="w-2 h-2 rounded-full bg-emerald-500 animate-pulse"></span>
                        Active Dev v2.2
                    </p>
                </div>
            </div>
            <nav class="hidden md:flex gap-6 text-sm font-medium text-slate-600">
                <a href="#pipeline" class="hover:text-blue-600 transition">Logic Pipeline</a>
                <a href="#architecture" class="hover:text-blue-600 transition">System Arch</a>
                <a href="#roadmap" class="hover:text-blue-600 transition">Roadmap</a>
                <a href="#setup" class="hover:text-blue-600 transition">Setup</a>
            </nav>
        </div>
    </header>

    <main class="flex-grow max-w-7xl mx-auto px-4 sm:px-6 lg:px-8 py-8 w-full space-y-12">

        <!-- Hero Section -->
        <section class="text-center space-y-4">
            <h2 class="text-4xl font-extrabold text-slate-900 tracking-tight sm:text-5xl">Programmatic Resale Automation</h2>
            <p class="max-w-2xl mx-auto text-lg text-slate-600">
                An enterprise-grade orchestration layer bridging unstructured visual data and structured e-commerce requirements using Hybrid AI.
            </p>
            <div class="flex justify-center gap-4 pt-4">
                <div class="px-4 py-2 bg-white rounded-full border border-slate-200 shadow-sm text-sm font-semibold text-slate-700 flex items-center gap-2">
                    <span class="text-blue-500">●</span> Locker Mode
                </div>
                <div class="px-4 py-2 bg-white rounded-full border border-slate-200 shadow-sm text-sm font-semibold text-slate-700 flex items-center gap-2">
                    <span class="text-emerald-500">●</span> Sourcing Mode
                </div>
                <div class="px-4 py-2 bg-white rounded-full border border-slate-200 shadow-sm text-sm font-semibold text-slate-700 flex items-center gap-2">
                    <span class="text-purple-500">●</span> Consignment
                </div>
                <div class="px-4 py-2 bg-white rounded-full border border-slate-200 shadow-sm text-sm font-semibold text-slate-700 flex items-center gap-2">
                    <span class="text-amber-500">●</span> Studio Mode
                </div>
            </div>
        </section>

        <!-- Section 1: The Logic Pipeline (Interactive) -->
        <section id="pipeline" class="scroll-mt-20">
            <div class="flex items-center justify-between mb-6">
                <div>
                    <h3 class="text-2xl font-bold text-slate-900">E-commerce Optimization Cycle</h3>
                    <p class="text-slate-500">Click on any stage to inspect the underlying service logic.</p>
                </div>
            </div>

            <!-- Pipeline Visual -->
            <div class="bg-white rounded-2xl shadow-sm border border-slate-200 p-8">
                <div class="grid grid-cols-1 md:grid-cols-5 gap-4 relative" id="pipeline-grid">
                    <!-- Nodes injected via JS -->
                </div>

                <!-- Detail View Panel -->
                <div id="pipeline-detail" class="mt-8 p-6 bg-slate-50 rounded-xl border border-slate-200 hidden transition-all">
                    <div class="flex flex-col md:flex-row gap-6">
                        <div class="md:w-1/3">
                            <h4 id="detail-title" class="text-xl font-bold text-blue-600 mb-2"></h4>
                            <div class="inline-flex items-center px-2.5 py-0.5 rounded-full text-xs font-medium bg-slate-200 text-slate-800 mb-4" id="detail-component"></div>
                            <p id="detail-desc" class="text-slate-600 text-sm leading-relaxed"></p>
                        </div>
                        <div class="md:w-2/3 bg-white rounded-lg border border-slate-200 p-4 shadow-inner">
                            <div class="flex justify-between items-center mb-2">
                                <span class="text-xs font-bold text-slate-400 uppercase tracking-wider">Key Functionality</span>
                            </div>
                            <ul id="detail-list" class="space-y-2 text-sm text-slate-700"></ul>
                        </div>
                    </div>
                </div>
            </div>
        </section>

        <!-- Section 2: Architecture & Roadmap -->
        <div class="grid grid-cols-1 lg:grid-cols-2 gap-8">

            <!-- Architecture Tree -->
            <section id="architecture" class="scroll-mt-20 bg-white rounded-2xl shadow-sm border border-slate-200 p-6 flex flex-col">
                <h3 class="text-xl font-bold text-slate-900 mb-4">System Architecture</h3>
                <p class="text-sm text-slate-500 mb-6">Flask Application Factory Pattern structure.</p>

                <div class="flex-grow overflow-auto bg-slate-50 rounded-xl p-4 border border-slate-200 font-mono text-sm text-slate-700">
                    <ul class="space-y-1" id="file-tree">
                        <!-- Tree injected via JS -->
                    </ul>
                </div>
            </section>

            <!-- Roadmap Status -->
            <section id="roadmap" class="scroll-mt-20 bg-white rounded-2xl shadow-sm border border-slate-200 p-6">
                <div class="flex justify-between items-center mb-6">
                    <h3 class="text-xl font-bold text-slate-900">Development Roadmap</h3>
                    <span class="px-2 py-1 bg-blue-100 text-blue-700 text-xs font-bold rounded-md">Phase 1 Active</span>
                </div>

                <div class="chart-container mb-6">
                    <canvas id="roadmapChart"></canvas>
                </div>

                <div class="space-y-4">
                    <div class="p-3 bg-slate-50 rounded-lg border border-slate-100">
                        <div class="flex justify-between items-center mb-2">
                            <h5 class="font-bold text-slate-800 text-sm">Phase 1: Automation</h5>
                            <span class="text-xs font-bold text-emerald-600">66% Complete</span>
                        </div>
                        <div class="w-full bg-slate-200 rounded-full h-1.5">
                            <div class="bg-emerald-500 h-1.5 rounded-full" style="width: 66%"></div>
                        </div>
                        <ul class="mt-2 space-y-1">
                            <li class="flex items-center gap-2 text-xs text-slate-500"><span class="text-emerald-500">✓</span> Project Refactor</li>
                            <li class="flex items-center gap-2 text-xs text-slate-500"><span class="text-emerald-500">✓</span> Business Policy Integration</li>
                            <li class="flex items-center gap-2 text-xs text-slate-500"><span class="text-slate-300">○</span> Return Window Lock</li>
                        </ul>
                    </div>
                </div>
            </section>
        </div>

        <!-- Section 3: Setup & Env -->
        <section id="setup" class="scroll-mt-20 bg-slate-900 rounded-2xl shadow-lg p-8 text-white">
            <h3 class="text-2xl font-bold mb-6 flex items-center gap-2">
                <svg class="w-6 h-6 text-emerald-400" fill="none" stroke="currentColor" viewBox="0 0 24 24"><path stroke-linecap="round" stroke-linejoin="round" stroke-width="2" d="M14 10l-2 1m0 0l-2-1m2 1v2.5M20 7l-2 1m2-1l-2-1m2 1v2.5M14 4l-2-1-2 1M4 7l2-1M4 7l2 1M4 7v2.5M12 21l-2-1m2 1l2-1m-2 1v-2.5M6 18l-2-1v-2.5M18 18l2-1v-2.5"></path></svg>
                Environment Initialization
            </h3>

            <div class="grid grid-cols-1 md:grid-cols-2 gap-8">
                <div>
                    <h4 class="text-sm font-bold text-slate-400 uppercase tracking-wider mb-3">Prerequisites</h4>
                    <ul class="space-y-2 text-sm text-slate-300">
                        <li class="flex items-center gap-2">📦 Python 3.14.2+ (Stable)</li>
                        <li class="flex items-center gap-2">🐳 Docker Desktop v4.37+</li>
                        <li class="flex items-center gap-2">🔑 eBay Developer Account (Sandbox)</li>
                        <li class="flex items-center gap-2">🧠 Google Gemini API Key</li>
                    </ul>
                </div>

                <div>
                    <h4 class="text-sm font-bold text-slate-400 uppercase tracking-wider mb-3">Quick Start</h4>
                    <div class="bg-black/50 rounded-lg p-4 font-mono text-xs text-emerald-400 border border-slate-700 relative group">
                        <p class="mb-1"><span class="text-slate-500"># 1. Clone & Install</span></p>
                        <p class="mb-2">git clone repo_url && pip install -r requirements.txt</p>
                        <p class="mb-1"><span class="text-slate-500"># 2. Config</span></p>
                        <p class="mb-2">cp .env.example .env</p>
                        <p class="mb-1"><span class="text-slate-500"># 3. Verify eBay Connect</span></p>
                        <p class="mb-2">python scripts/ebay_inventory_check.py</p>
                        <p class="mb-1"><span class="text-slate-500"># 4. Launch</span></p>
                        <p>python run.py</p>

                        <button onclick="copySetup()" class="absolute top-2 right-2 p-1.5 bg-slate-700 hover:bg-slate-600 rounded text-white opacity-0 group-hover:opacity-100 transition">
                            <svg class="w-4 h-4" fill="none" stroke="currentColor" viewBox="0 0 24 24"><path stroke-linecap="round" stroke-linejoin="round" stroke-width="2" d="M8 16H6a2 2 0 01-2-2V6a2 2 0 012-2h8a2 2 0 012 2v2m-6 12h8a2 2 0 002-2v-8a2 2 0 00-2-2h-8a2 2 0 00-2 2v8a2 2 0 002 2z"></path></svg>
                        </button>
                    </div>
                </div>
            </div>
        </section>

    </main>

    <footer class="bg-white border-t border-slate-200 mt-auto">
        <div class="max-w-7xl mx-auto px-4 sm:px-6 lg:px-8 py-8 flex flex-col md:flex-row justify-between items-center">
            <p class="text-sm text-slate-500">© 2026 AI List Assist Project.</p>
            <div class="flex gap-6 text-sm text-slate-500">
                <span class="hover:text-blue-600 cursor-pointer">Strategy Guide</span>
                <span class="hover:text-blue-600 cursor-pointer">Funding Arch</span>
                <span class="hover:text-blue-600 cursor-pointer">MSP Docs</span>
            </div>
        </div>
    </footer>

    <!-- JavaScript Logic -->
    <script>
        // --- Data Models ---
        const pipelineData = [
            {
                id: 'visual',
                title: 'Visual Acquisition',
                component: 'VisionService',
                desc: 'Utilizes Hybrid Vision (OCR + Generative AI) to detect items, read labels, and extract product specs from a single image.',
                features: ['Object Detection', 'Text OCR', 'Condition Assessment', 'Image Quality Check']
            },
            {
                id: 'tax',
                title: 'Taxonomy Validation',
                component: 'EBayCategoryService',
                desc: 'Maps raw visual data to strict category-specific rules defined by eBay taxonomy to ensure listing compliance.',
                features: ['Category ID Mapping', 'Aspect Constraint Check', 'Required Field ID']
            },
            {
                id: 'market',
                title: 'Market Analysis',
                component: 'ValuationService',
                desc: 'Queries real-time sold listings to determine value. Filters unprofitable items via the Decision Gate logic.',
                features: ['Sold Listings Query', 'Profit Calculation', 'Price Floor Logic', 'Sell-Through Rate']
            },
            {
                id: 'convo',
                title: 'Conversational Orchestrator',
                component: 'ConversationOrchestrator',
                desc: 'State machine that identifies missing fields and generates targeted user queries to resolve ambiguities.',
                features: ['Progressive Questioning', 'State Management', 'Ambiguity Resolution', 'User Intent Gate']
            },
            {
                id: 'api',
                title: 'API Synthesis',
                component: 'ListingSynthesisEngine',
                desc: 'Compiles the ListingDraft into Inventory Item & Offer objects, then pushes them to the eBay Sell API.',
                features: ['Inventory Item Creation', 'Offer Generation', 'Policy Application', 'Publish Trigger']
            }
        ];

        const fileTreeData = [
            { name: 'app/', type: 'folder', children: [
                { name: '__init__.py', type: 'file', note: 'Factory & Config' },
                { name: 'routes.py', type: 'file', note: 'API Endpoints' },
                { name: 'templates/', type: 'folder', note: 'Dashboards' }
            ]},
            { name: 'services/', type: 'folder', children: [
                { name: 'vision_service.py', type: 'file' },
                { name: 'valuation_database.py', type: 'file' },
                { name: 'conversation_orchestrator.py', type: 'file' },
                { name: 'ebay_policy_service.py', type: 'file' }
            ]},
            { name: 'scripts/', type: 'folder', children: [
                { name: 'get_token.py', type: 'file' },
                { name: 'ebay_inventory_check.py', type: 'file' }
            ]},
            { name: 'docs/', type: 'folder', note: 'Architecture & Prompts' },
            { name: 'config.py', type: 'file', note: 'Env Vars' },
            { name: 'run.py', type: 'file', note: 'Entry Point' }
        ];

        // --- Render Logic ---

        // 1. Pipeline Renderer
        function renderPipeline() {
            const container = document.getElementById('pipeline-grid');
            container.innerHTML = pipelineData.map((node, index) => `
                <div onclick="selectPipelineNode(${index})"
                     class="cursor-pointer p-4 rounded-xl border-2 border-slate-200 hover:border-blue-400 hover:shadow-md transition-all bg-white relative group ${index === 0 ? 'active-node' : ''}"
                     id="node-${index}">
                    <div class="flex items-center justify-center w-8 h-8 rounded-full bg-slate-100 text-slate-500 font-bold text-sm mb-3 group-hover:bg-blue-100 group-hover:text-blue-600 transition">
                        ${index + 1}
                    </div>
                    <h4 class="font-bold text-slate-800 text-sm mb-1">${node.title}</h4>
                    <p class="text-xs text-slate-500 font-mono">${node.component}</p>
                    ${index < pipelineData.length - 1 ? `
                        <div class="hidden md:block absolute -right-3 top-1/2 -translate-y-1/2 z-10">
                            <svg class="w-6 h-6 text-slate-300" fill="none" stroke="currentColor" viewBox="0 0 24 24"><path stroke-linecap="round" stroke-linejoin="round" stroke-width="2" d="M9 5l7 7-7 7"></path></svg>
                        </div>
                    ` : ''}
                </div>
            `).join('');

            // Initialize with first node
            selectPipelineNode(0);
        }

        function selectPipelineNode(index) {
            // Update Styles
            document.querySelectorAll('#pipeline-grid > div').forEach((el, i) => {
                if (i === index) el.classList.add('active-node', 'border-blue-500');
                else el.classList.remove('active-node', 'border-blue-500');
            });

            // Update Content
            const data = pipelineData[index];
            const detailPanel = document.getElementById('pipeline-detail');

            detailPanel.classList.remove('hidden');
            document.getElementById('detail-title').textContent = data.title;
            document.getElementById('detail-component').textContent = data.component;
            document.getElementById('detail-desc').textContent = data.desc;

            const list = document.getElementById('detail-list');
            list.innerHTML = data.features.map(f => `
                <li class="flex items-center gap-2">
                    <svg class="w-4 h-4 text-emerald-500" fill="none" stroke="currentColor" viewBox="0 0 24 24"><path stroke-linecap="round" stroke-linejoin="round" stroke-width="2" d="M5 13l4 4L19 7"></path></svg>
                    ${f}
                </li>
            `).join('');
        }

        // 2. File Tree Renderer
        function renderTree(nodes, container, level = 0) {
            nodes.forEach(node => {
                const li = document.createElement('li');
                li.className = `pl-${level * 4} flex items-center py-1 hover:bg-slate-100 rounded`;

                const icon = node.type === 'folder'
                    ? `<svg class="w-4 h-4 text-blue-400 mr-2" fill="currentColor" viewBox="0 0 20 20"><path d="M2 6a2 2 0 012-2h5l2 2h5a2 2 0 012 2v6a2 2 0 01-2 2H4a2 2 0 01-2-2V6z"></path></svg>`
                    : `<svg class="w-4 h-4 text-slate-400 mr-2" fill="none" stroke="currentColor" viewBox="0 0 24 24"><path stroke-linecap="round" stroke-linejoin="round" stroke-width="2" d="M9 12h6m-6 4h6m2 5H7a2 2 0 01-2-2V5a2 2 0 012-2h5.586a1 1 0 01.707.293l5.414 5.414a1 1 0 01.293.707V19a2 2 0 01-2 2z"></path></svg>`;

                li.innerHTML = `${icon} <span class="${node.type === 'folder' ? 'font-bold text-slate-800' : 'text-slate-600'}">${node.name}</span> ${node.note ? `<span class="ml-2 text-xs text-slate-400">// ${node.note}</span>` : ''}`;
                container.appendChild(li);

                if (node.children) {
                    renderTree(node.children, container, level + 1);
                }
            });
        }

        // 3. Chart Renderer
        function initChart() {
            const ctx = document.getElementById('roadmapChart').getContext('2d');
            new Chart(ctx, {
                type: 'bar',
                data: {
                    labels: ['Automation', 'Reporting', 'Scale'],
                    datasets: [{
                        label: 'Phase Completion %',
                        data: [66, 0, 0],
                        backgroundColor: ['#10B981', '#E2E8F0', '#E2E8F0'],
                        borderRadius: 6,
                        barThickness: 40
                    }]
                },
                options: {
                    responsive: true,
                    maintainAspectRatio: false,
                    plugins: { legend: { display: false } },
                    scales: {
                        y: { beginAtZero: true, max: 100, grid: { display: false } },
                        x: { grid: { display: false } }
                    }
                }
            });
        }

        // 4. Utilities
        function copySetup() {
            const code = `git clone repo_url && pip install -r requirements.txt\ncp .env.example .env\npython scripts/ebay_inventory_check.py\npython run.py`;
            navigator.clipboard.writeText(code);
            alert('Setup commands copied to clipboard!');
        }

        // Initialization
        document.addEventListener('DOMContentLoaded', () => {
            renderPipeline();
            renderTree(fileTreeData, document.getElementById('file-tree'));
            initChart();
        });

    </script>
</body>
</html>
```
